```python
from PIL import Image, ImageDraw
import yaml
```

```python
img_path = 'output/data_de/0.png'
metadata_path = 'output/data.de'

img = Image.open(img_path)
with open(metadata_path, 'r') as f:
    metadata = yaml.safe_load(f)
bboxes = metadata[0]['bboxes']
```

```python
img_copy = img.copy()
draw = ImageDraw.Draw(img_copy)

for bbox in bboxes:
    # bbox is a list of 4 points: [[x1,y1],[x2,y2],[x3,y3],[x4,y4]]
    # Convert to flat list for polygon drawing
    points = [tuple(point) for point in bbox]
    draw.polygon(points, outline='red', width=2)

img_copy.show()
```
